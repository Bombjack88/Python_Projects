- TV, Jornal e Rádio estão em milhares de reais

```python
# importar a base de dados

import pandas as pd

tabela=pd.read_csv("advertising.csv")
display(tabela)

# tratamento de dados
print(tabela.info())
```

#### Análise Exploratória
- Vamos tentar visualizar como as informações de cada item estão distribuídas
- Vamos ver a correlação entre cada um dos itens

```python
# 3 bibliotecas de análise gráfica
#plotly
#matplotlib
#seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# criar o gráfico

sns.heatmap(tabela.corr(),cmap="Blues", annot=True)
# exibi o gráfico
plt.show()
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
y=tabela["Vendas"] #quem eu quero prever
x=tabela[["TV","Radio","Jornal"]] #quem eu vou usar para fazer a previsão
```

```python
#separar os dados em treino e teste (o padrão é 80 treino e 20 teste, se quiser alterar adiciono o parametro test_size)
from sklearn.model_selection import train_test_split # existe uma função que faz a separação

x_treino, x_teste, y_treino, y_teste=train_test_split(x,y, test_size=0.3) # a ordem tem que ser sempre esta (foi o padrão que foi criado)
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

#criou a ineligencia artificial
modelo_regressaolinear=LinearRegression()
modelo_arvoredecisao=RandomForestRegressor(n_estimators=10)

# treinou a inteligencia artificail
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
# qual é o melhor modelo (quem ganha o jogo)

previsao_regressaolinear=modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao=modelo_arvoredecisao.predict(x_teste)

from sklearn import metrics

print(metrics.r2_score(y_teste,previsao_regressaolinear))
print(metrics.r2_score(y_teste,previsao_arvoredecisao))


# DS: alternativa - Acuracia
resultado1=modelo_regressaolinear.score(x_teste,y_teste)
print("Acurácia Regressão Linear:", resultado1)

resultado2=modelo_arvoredecisao.score(x_teste,y_teste)
print("Acurácia Regressão Linear:", resultado2)
```

#### Visualização Gráfica das Previsões

```python
# a árvore de decisão é o melhor modelo
tabela_auxiliar=pd.DataFrame()
tabela_auxiliar['y_teste']=y_teste
tabela_auxiliar['Previsão Regressão linear']=previsao_regressaolinear
tabela_auxiliar['Previsão àrvore de Decisão']=previsao_arvoredecisao

plt.figure(figsize=(10,6))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```
