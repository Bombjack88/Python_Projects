# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#!pip install pyautogui
#!pip install pyperclip
```

```python
# Passo a passo para resolver o nosso desafio

import pyautogui
import pyperclip # auxiliar que consegue trabalhar com simbolos ou acentos
import time

#tempo/delay
pyautogui.PAUSE = 1


# Passo1: Entrar no sistema da empresa (link do drive)
## DS: Abrir o navegador
pyautogui.hotkey("ctrl","t")

## DS: Escreve o endereço - pode não fucncionar devido aos caracteres especiais (no meu fucniciona)

#pyautogui.write("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing") #escreve o endereço

## DS: Alternativa - Utilizo o pyperclip para copiar o endereço porque tinha um caractere especial depois vou colar utilizando o pyautogui
## DS: Copiar o link
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
## DS: colar o link
pyautogui.hotkey("ctrl", "v")

## DS: fazer enter - entra no sistema
pyautogui.press("enter")

## DS: problema: pode demorar a carregar a página - por exemplo quer dar 5 segundos
time.sleep(5)

# Passo2 : Navegar no sistema (até a pasta exportar)
##DS: saber a posição e fazer duplo click
pyautogui.click(x=285, y=254,clicks=2)
time.sleep(2)

# Passo 3: Fazer o download da base de vendas
## DS: são 3 clickes
pyautogui.click(x=345, y=254) # clica no ficheiro
pyautogui.click(x=2355, y=163) #clica nos 3 pontinhos
pyautogui.click(x=2101, y=528) #clica no download
time.sleep(5)

```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Importar a base de vendas para o Python

## DS: Se não estiver a utilizar o Jupyper poreciso de instalar
# pandas
# numpy
# openpyxl

import pandas as pd

tabela=pd.read_excel(r"C:\Users\fiu126\Downloads\Vendas - Dez.xlsx",sheets=1)
display(tabela)

# Passo 5: Calcular o faturamento e quantidade de produtos vendidos(os indicadores)
faturamento= tabela["Valor Final"].sum()
qtde_produtos=tabela["Quantidade"].sum()
display(faturamento)
display(qtde_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 6: Enviar email para diretoria
link="https://outlook.live.com/mail/0/"

# DS: abrir o e-mail
pyautogui.hotkey("ctrl","t")
pyperclip.copy("https://outlook.live.com/mail/0/")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")
time.sleep(5)

# DS: clicar no botão escrever
pyautogui.click(x=260, y=159)
# DS: preencher o destino
#pyautogui.write("cash-flow@hotmail.com")

pyperclip.copy("cash-flow@hotmail.com")
pyautogui.hotkey("ctrl","v")

time.sleep(2)
#pyautogui.press("tab") #seleciona o e-mail

pyautogui.press("tab") #muda para o campo de assunto

# DS: preencher o assunto
#pyautogui.write("Relatório de Vendas")
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl","v")

pyautogui.press("tab") # muda para o corpo do e-mail


# DS: escrever o e-mail

texto= f"""
Prezamos, bom dia 

O faturamento de ontem foi de R$:{faturamento:,.2f}.
A quantidade de produtos foi de: {qtde_produtos:,}.

Abraço

Daniel Python
"""
pyperclip.copy(texto)

pyautogui.hotkey("ctrl","v")

# DS :enviar

pyautogui.hotkey("ctrl","enter")
```
